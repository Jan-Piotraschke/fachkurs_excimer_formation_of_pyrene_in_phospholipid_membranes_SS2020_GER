```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
testPd = pd.read_csv('data/Pyren-DPPC_Device4_T.csv', header=None, sep=';',index_col=0)
#testPd = pd.read_csv('data/Pyren-EiPC_Device3_T.csv', header=None, sep='\t',index_col=0)
testPd.columns = testPd.iloc[0]
testPd = testPd.drop(testPd.index[0])
newPd = pd.DataFrame(np.array(testPd).astype('float64'), index=testPd.index.tolist(), columns=testPd.columns.tolist())

del newPd['49C']
del newPd['51C']
```
