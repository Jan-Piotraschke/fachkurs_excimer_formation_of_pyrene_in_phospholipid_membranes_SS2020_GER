```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

!pip install autoviz
```

```python
devices = ["Device3", "Device4"]
selectedDevice = devices[0]
```

```python
# testPd = pd.read_csv('data/Pyren-EiPC_Device3_c.csv', header=None, sep=';',index_col=0)
testPd = pd.read_csv('data/Pyren-EiPC_{}_c.csv'.format(selectedDevice), header=None, sep=';',index_col=0)

testPd.columns = testPd.iloc[0]
testPd = testPd.drop(testPd.index[0])
newPd = pd.DataFrame(np.array(testPd).astype('float64'), index=testPd.index.tolist(), columns=testPd.columns.tolist())
```

```python
# Importing Autoviz class from the autoviz library
from autoviz.AutoViz_Class import AutoViz_Class
#Initialize the Autoviz class in a object called df
df = AutoViz_Class()
# Using Iris Dataset and passing to the default parameters
filename = 'data/Pyren-EiPC_{}_c.csv'.format(selectedDevice)
sep = ";"
graph = df.AutoViz(
    filename,
    sep=";",
    depVar="",
    dfte=None,
    header=0,
    verbose=0,
    lowess=False,
    chart_format="svg",
    max_rows_analyzed=150000,
    max_cols_analyzed=30,
)
```

```python
plt.figure(figsize=(8,4))

ax = plt.axes()
ax.set_xlabel('Wellenlänge $\lambda$ [nm]')
ax.set_ylabel('Fluoreszensintensität [a.u.]')

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.legend()
ax.yaxis.grid(alpha=0.4)

for i in newPd.columns.tolist():
  plt.plot(newPd[i],'-',label=i)

plt.xlim(360, 500)  
plt.ylim(0)  

plt.legend(frameon=False)

plt.savefig('reports/Konz_{}.png'.format(selectedDevice),dpi=600, bbox_inches='tight')

plt.show()
```

*Hinweis*: die beiden Plots der Geräte unterscheiden sich zwischen 50:1 und 200:1 wie als wären sie vertauscht, aber nach den Metadaten sind die Daten in der korrekten Reihenfolge
